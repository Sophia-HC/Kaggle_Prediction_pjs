```python
import pandas as pd
train_df = pd.read_csv("tatinic_train.csv", encoding="utf-8")
train_df
```

```python
import pandas as pd
test_df = pd.read_csv("tatinic_test.csv", encoding="utf-8")
test_df
```

```python
                 
train_df.isna().sum()                  
```

```python
med = train_df.median()                      
train_df = train_df.fillna(med)              
train_df.isna().sum()                        

#train_df.median()                          
```

```python
test_df.isna().sum()
```

```python
test_df = test_df.fillna(med)               
test_df.isna().sum()                        
```

```python
most = train_df['Embarked'].value_counts().idxmax()      
train_df['Embarked'] = train_df['Embarked'].fillna(most)  
train_df.isna().sum()                                
```

```python
test_df['Embarked'] = test_df['Embarked'].fillna(most)   
test_df.isna().sum()                                     
```

```python
dummy = pd.get_dummies(train_df['Embarked'])                  
        
train_df = pd.concat([train_df, dummy], axis=1)                #pd.concat用法: ttps://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.concat.html
train_df
```

```python
dummy = pd.get_dummies(test_df['Embarked'])                   
         
test_df = pd.concat([test_df, dummy], axis=1)               
test_df
```

```python

dummy = pd.get_dummies(train_df['Sex'])                        
train_df = pd.concat([train_df, dummy], axis=1) 
train_df

dummy = pd.get_dummies(test_df['Sex'])                        
test_df = pd.concat([test_df, dummy], axis=1) 
test_df
```

```python

def nameflow(s):                              
    mid = s.split(",")[-1].split(".")[0]           
    mid = mid.replace(" ","")                  
    if mid == 'Miss':
        return 'Miss'
    elif mid == 'Mr':
        return 'Mr'
    elif mid == 'Mrs':
        return 'Mrs'
    else:
        return 'Throw'        
name = train_df['Name'].apply(nameflow)

dummy = pd.get_dummies(name)              
          
train_df = pd.concat([train_df, dummy], axis=1)               
train_df
```

```python

name = test_df['Name'].apply(nameflow)

dummy = pd.get_dummies(name)                
test_df = pd.concat([test_df, dummy], axis=1)               
test_df
```

```python

train_df = train_df.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'Throw'], axis = 1) 
test_df = test_df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'Throw'], axis = 1)   
train_df
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np                                            

trainx = train_df.drop(['Survived'], axis=1)                 
trainy = train_df['Survived']
clf = RandomForestClassifier(max_depth=7, n_estimators=22)                                                               
np.average(cross_val_score(clf, trainx, trainy, cv=10))                     
```

```python

from sklearn.model_selection import GridSearchCV
p = {
    "max_depth":range(5, 26),
    "n_estimators":range(20, 100)
}
clf = RandomForestClassifier()
s = GridSearchCV(clf, p, cv=10)
s.fit(trainx, trainy)
print(s.best_params_)            
print(s.best_score_)         
```

```python
testx = test_df.drop(["PassengerId"], axis=1)                    
testid = test_df["PassengerId"]
clf = RandomForestClassifier(max_depth=9, n_estimators=25)       
clf.fit(trainx, trainy)                                          
pre = clf.predict(testx)                                        

result = pd.DataFrame()                                          
result["PassengerId"] = testid                                   
result["Survived"] = pre                                         

# result.to_csv("tatinic_result.csv", encoding="utf-8", index=False)       
result
```
